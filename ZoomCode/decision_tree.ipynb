```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier as DTS
from sklearn.metrics import accuracy_score as score 

```

```python
data = np.asarray(pd.read_csv('data.csv', header=None))
data
```

```python
x = data[:,0:2]
y = data[:,2]
print("x:==>", x, "\n", "y:==>", y)
```

```python
model = DTS()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
prediction = list(zip(y, y_pred))
prediction
```

```python
from itertools import count
from operator import le


TruePredCount = 0
for a,b in prediction:
    if not a == b:
        print(prediction) 
    else:
       TruePredCount += 1
print(TruePredCount, len(x))

        
```

```python
print(score(y, y_pred))
```
